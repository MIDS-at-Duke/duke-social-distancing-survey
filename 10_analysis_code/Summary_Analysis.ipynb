```python
svy = pd.read_csv('../00_raw_data/20200401_duke_covid_survey/raw_survey_data_uniqueid_only_duke_202004.txt', 
                  delimiter='\t')

for i in codes.keys():
    if i in svy.columns and i not in ['q4', 'q5','q6', 'q7', 'q10']:
        svy[i] = svy[i].astype('str').str.replace(".0", "")
        svy[i] = svy[i].replace('nan', np.nan)
        svy[i] = svy[i].replace(codes[i])
        
drop_rollups = [n for n in svy.columns if not re.match('.*_rollup', n)]
svy = svy[drop_rollups]
svy = svy.rename(names, axis='columns')
svy.sample().T
```
